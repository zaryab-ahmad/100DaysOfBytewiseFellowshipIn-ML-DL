```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

## Load Dataset
load the minst but this dataset is very popular so the Keras is store in own library.
MNIST dataset is the collection of handwritten digits (0-9).

```python
(X_train, y_train),(X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_test.shape
```

```python
X_train[1]
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[500]) # last number of training data
```

```python
X_train = X_train/255  # to normalize the image the pixel values are scale to a range  between 0 and 1
X_test = X_test/255
```

```python
X_train[0]
```

## Creating the Layers

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

## Model Compile

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer='Adam',metrics=['accuracy'])
```

# Training the Model

```python
history = model.fit(X_train,y_train, epochs=5, validation_split=0.2 )
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis = 1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
