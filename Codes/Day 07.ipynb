<h4>Exercise:</h4> Create a simple line plot using Matplotlib to display the trend of a list of values over time.

```python
import matplotlib.pyplot as plt

# Sample data
time = ['2024-07-01', '2024-07-02', '2024-07-03', '2024-07-04', '2024-07-05']
values = [10, 15, 13, 18, 16]

# Convert the time data to a format suitable for plotting
time_points = range(len(time))

# Create the line plot
plt.plot(time_points, values, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xticks(time_points, time)
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Trend of Values Over Time')

# Show the plot
plt.show()
```

<h4>Exercise:</h4> Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset.`m

```python
corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Tips Dataset')
plt.show()
```

<h4>Exercise: </h4>Use Matplotlib to create a subplot grid that displays multiple charts in a single figure.


```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))


sns.countplot(x='Survived', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Survival Count')


sns.countplot(x='Pclass', data=df, ax=axs[0, 1])
axs[0, 1].set_title('Passenger Class Distribution')


sns.histplot(df['Age'].dropna(), bins=20, kde=True, ax=axs[1, 0])
axs[1, 0].set_title('Age Distribution')


sns.histplot(df['Fare'], bins=20, kde=True, ax=axs[1, 1])
axs[1, 1].set_title('Fare Distribution')


plt.tight_layout()
plt.show()
```

<h4>Exercise:</h4> Create a violin plot using Seaborn to visualize the distribution of a dataset across different categories.

```python
sns.set(style='whitegrid')

plt.figure(figsize=(10, 6))
sns.violinplot(x='Pclass', y='Age', data=df, palette='Set2')

plt.title('Distribution of Passenger Ages Across Classes', fontsize=16)
plt.xlabel('Passenger Class', fontsize=14)
plt.ylabel('Age', fontsize=14)

plt.show()
```

<h4>Exercise: </h4>Combine Matplotlib and Seaborn to create a complex visualization, such as overlaying a KDE plot on a histogram.

```python
age_data = df['Age'].dropna()


sns.set(style='whitegrid')


plt.figure(figsize=(12, 6))


sns.histplot(age_data, kde=True, bins=20, color='skyblue', edgecolor='blue', linewidth=1.5)


plt.title('Distribution of Passenger Ages with KDE Overlay', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)


plt.show()
```
