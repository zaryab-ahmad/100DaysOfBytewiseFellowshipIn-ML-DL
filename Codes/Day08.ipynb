#### Exercise:
Calculate the mean, median, and mode of the sepal lengths in the Iris dataset.



## Importing the libraries 

```python
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Loading the Iris Dataset

```python
iris = sns.load_dataset('iris')
```

### Calculate mean, median, and mode of sepal lengths

```python
sepal_lengths = iris['sepal_length']
mean_sepal_length = sepal_lengths.mean()
median_sepal_length = sepal_lengths.median()
mode_sepal_length = stats.mode(sepal_lengths)[0][0]
```

### Print the values of [Mean, Median, and mode]

```python
print(f"Mean of sepal lengths: {mean_sepal_length}")
print(f"Median of sepal lengths: {median_sepal_length}")
print(f"Mode of sepal lengths: {mode_sepal_length}")
```

### Calculate variance and standard deviation of petal widths

```python
petal_widths = iris['petal_width']
variance_petal_width = petal_widths.var()
std_dev_petal_width = petal_widths.std()
```

## Printing the value of the [variance, standard deviation of petal widthd]

```python
print(f"Variance of petal widths: {variance_petal_width}")
print(f"Standard deviation of petal widths: {std_dev_petal_width}")
```

##### Exercise:
Calculate the cumulative distribution function (CDF) for the petal lengths and plot it.

#### Extract the petal lengths

```python
petal_lengths = iris['petal_length']
```

#### Sort the petal lengths

```python
sorted_petal_lengths = np.sort(petal_lengths)
```

```python
cdf_values = np.arange(1, len(sorted_petal_lengths) + 1) / len(sorted_petal_lengths)
```

#### Plot the CDF

```python
plt.figure(figsize=(10, 6))
plt.plot(sorted_petal_lengths, cdf_values, marker='.', linestyle='none')
plt.title('Cumulative Distribution Function (CDF) of Petal Lengths')
plt.xlabel('Petal Length (cm)')
plt.ylabel('CDF')
plt.show()
```

##### Exercise: 
Determine the probability of a randomly selected iris flower having a petal length greater than a given value.

#### Extract the petal lengths

```python
petal_lengths = iris['petal_length']
```

#### Function to calculate the probability

```python
def probability_greater_than(value):
    sorted_petal_lengths = np.sort(petal_lengths)
    cdf_values = np.arange(1, len(sorted_petal_lengths) + 1) / len(sorted_petal_lengths)
    prob_less_than_or_equal = np.searchsorted(sorted_petal_lengths, value, side='right') / len(sorted_petal_lengths)
    prob_greater_than = 1 - prob_less_than_or_equal
    return prob_greater_than
```

```python
given_value = 4.5
probability = probability_greater_than(given_value)
```

```python
print(f"Probability of a randomly selected iris flower having a petal length greater than {given_value} cm: {probability}")
```

#### Plot the CDF for reference

```python
plt.figure(figsize=(10, 6))
sorted_petal_lengths = np.sort(petal_lengths)
cdf_values = np.arange(1, len(sorted_petal_lengths) + 1) / len(sorted_petal_lengths)
plt.plot(sorted_petal_lengths, cdf_values, marker='.', linestyle='none')
plt.axvline(x=given_value, color='r', linestyle='--', label=f'Petal Length = {given_value}')
plt.title('Cumulative Distribution Function (CDF) of Petal Lengths')
plt.xlabel('Petal Length (cm)')
plt.ylabel('CDF')
plt.legend()
plt.show()
```

##### Exercise:
Perform a hypothesis test to determine if there is a significant difference in the mean petal length between two species of iris flowers.

##### Extract petal lengths for the two species

```python
setosa_petal_lengths = iris[iris['species'] == 'setosa']['petal_length']
versicolor_petal_lengths = iris[iris['species'] == 'versicolor']['petal_length']
```

##### Perform the two-sample t-test

```python
t_statistic, p_value = stats.ttest_ind(setosa_petal_lengths, versicolor_petal_lengths)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```

#### Determine if there is a significant difference

```python
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in the mean petal length between setosa and versicolor.")
else:
    print("There is no significant difference in the mean petal length between setosa and versicolor.")
```

##### Exercise:
Calculate and interpret the covariance and correlation between sepal length and sepal width.


##### Extract sepal length and sepal width

```python
sepal_length = iris['sepal_length']
sepal_width = iris['sepal_width']
```

###### Calculate covariance and correlation

```python
covariance = sepal_length.cov(sepal_width)

correlation = sepal_length.corr(sepal_width)
```

```python
print(f"Covariance between sepal length and sepal width: {covariance}")
print(f"Correlation between sepal length and sepal width: {correlation}")
```
